```python
import numpy as np
```

```python
# create a 1D array
x1 = np.array([3, 4, 5])
x1
```

```python
# sum arrays
x2 = np.array([6, 7, 8])
x1 + x2
```

```python
# create a 2D array
x = np.array([[1, 2],
              [3, 4]])
x
```

```python
# getting dimensions, shape, size, and data type of the array
x.ndim, x.shape, x.size, x.dtype
```

```python
# sum of all elements
x.sum()
```

```python
# reshaping the array (showing that numpy arrays are row-major ordering)
x =np.array([1, 2, 3, 4, 5, 6])
print("beginning x:\n", x)
x_reshaped = x.reshape(2, 3)
print("reshaped x:\n", x_reshaped)
```

```python
# accessing elements in the reshaped array
print(x_reshaped[1, 2])
print(x_reshaped[0])
```

```python
# showing that numpy arrays copies are shallow
print("before modifying x_reshaped:\n", x_reshaped)
x_reshaped[0, 0] = 5
print("x_reshaped after modifying:\n", x_reshaped)
print("x after modifying x_reshaped:\n", x)
```

```python
# transpose
x_transposed = x_reshaped.T
print("x_reshaped transposed:\n", x_transposed)
```

```python
# square root of each element
print("Using sqrt():\n", np.sqrt(x))
print("Using ** 0.5:\n", x ** 0.5)
```

```python
# random numbers from a normal distribution
# loc=mean, scale=std_dev, size=number of samples
x = np.random.normal(size=50)
```

```python
# new array created by adding independent N(50, 1)
y = x + np.random.normal(loc=50, scale=1, size=50)
y
```

```python
# correlation matrix between x and y
# off-diagonal elements are the correlation coefficients
correlation_matrix = np.corrcoef(x, y)
print("Correlation matrix:\n", correlation_matrix)
```

```python
# set random seed for reproducibility
rng = np.random.default_rng(1303)
rng.normal(size=10)
```

```python
# functions to compute mean, median, variance, and standard deviation
rng = np.random.default_rng(3)
y = rng.standard_normal(10)
print(f"y:\n {y}", 
f"\nMean:\n {np.mean(y):.3f}", 
f"\nMedian:\n {np.median(y):.3f}", 
f"\nVariance:\n {np.var(y, ddof=1):.3f}", 
f"\nStandard dev:\n {np.std(y, ddof=1):.3f}")
```

```python
# applying functions to part of the array
rng = np.random.default_rng(3)
x = rng.standard_normal((10, 3))
print("x:\n", x)
print("\nMean:\n", x.mean(axis=0)) # mean of each column
```

### 2.3.6 Indexing Data

```python
# arange creates an array with evenly spaced values within a given interval
A = np.array(np.arange(16)).reshape(4, 4)
A
```

```python
# Retrieve second and fourth rows
A[[1, 3]]
```

```python
# Retrieve second and fourth columns
A[:, [1, 3]]
```

```python
# Getting a two elements (number 4 and 14)
A[[1, 3], [0, 2]]
```

```python
# Creating submatrix of row 1 and 3, and columns 0 and 2
A[[1, 3]][:,[0, 2]] 
```

```python
# Same as above, but using a different syntax
idx = np.ix_([1, 3], [0, 2])
A[idx]
```

```python
# Or alternatively using slices [first item : last item : step]
A[1:4:2, 0:3:2]  # rows 1 and 3, columns 0 and 2
```

Boolean Indexing

```python
# Create a vector of 0's, lenght equal to the first dimension of A
keep_rows = np.zeros(A.shape[0], dtype=bool)
keep_rows
```

```python
# Set two elements to True
keep_rows[[1, 3]] = True
keep_rows
```

```python
np.all(keep_rows == np.array([False, True, False, True]))
```

```python
# Integers and Booleans are treated differently in numpy
# Boolean indexing
A[np.array([False, True, False, True])]
```

```python
# Integer indexing
A[np.array([0, 1, 0, 1])]
```

```python
 # Using boolean indexing with multiple dimensions
 keep_cols = np.zeros(A.shape[1], bool)
 keep_cols[[0, 2, 3]] = True
 idx_bool = np.ix_(keep_rows, keep_cols)
 A[idx_bool]
```

```python
# Create DataFrame with 20% missing values
rng = np.random.default_rng(1)
A = rng.standard_normal((127, 5))
M = rng.choice([0, np.nan], size=A.shape, p=[0.8, 0.2])
A += M
D = pd.DataFrame(A, columns=['food', 'bar', 'pickle', 'snack', 'popcorn'])
D 
```

```python
for col in D.columns:
    template = 'Column "{0}" has {1:.2f}% missing values'
    print(template.format(col, D[col].isna().mean() * 100))
```
